A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden:
latitude
longitude
name
label
Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.

```python
import numpy as np
import pandas as pd

```

```python
df = pd.read_csv("hospitales2.csv", sep=',') #Cargo la data

df.head() #muestro la estructura
```

```python
df_new = df.loc[:,['WKT', 'DOM_NORMA', 'NOMBRE']] #creo un nuevo DF con los datos que se van a usar
```

```python
#quiro los caracteres que me sobran en el campo (POINT Y parentesis
df_new['WKT'] = df_new['WKT'].str.replace("POINT", "", regex=False).str.replace("(", "", regex=False).str.replace(")", "", regex=False)
```

```python
#separo los datos de longitud y latitud en dos columnas nuevas
df_new = df_new.assign(latitude=df_new['WKT'].str.split().str[0], longitude=df_new["WKT"].str.split().str[1])
```

```python
df_new = df_new.drop('WKT', axis=1) #Elimino la columa que ya no uso
```

```python
df_new = df_new[['latitude', 'longitude', 'DOM_NORMA', "NOMBRE",]] #Reordeno las columnas
```

```python
#renombrar las columnas
df_new = df_new.rename(columns={"DOM_NORMA": "nombre", "NOMBRE": "label"})
```

```python
df_new.to_csv("hospital_hwk_ok.csv", index=False)
```
